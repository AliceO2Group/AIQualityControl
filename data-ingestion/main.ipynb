# Build the book-keeping API URL, and use it to fetch the Runs that satisfy the desired filters (ex. stable beams)

```python
BKKP_RUN_API_URL = build_bkkp_run_api_url(
        runs_limit=5000,
        detector="TPC",
        run_definition="PHYSICS",
        run_qualities="good",
        personal_token=PERSONAL_TOKEN,
        tag_value="Not+for+physics",
        tag_operation="none-of"
    )

custom_filters = [
        is_in_stable_beams,
        #lambda run: has_beam_type(run, "PP"),
        lambda run: has_good_detector_quality(run, "TPC"),
    ]

fetch_and_save_runs(url=BKKP_RUN_API_URL,
                    raw_output_path="runs_raw.json", 
                    stable_output_path="bkkp_data/runs_stable_beams_with_good_tpc_quality_auto.json", 
                    filters=custom_filters,
                    )
```
