# Compare images and tensors

```python
import os
import numpy as np
from PIL import Image

img_folder = "/Users/zetasourpi/cernbox/data/data-for-ml/good_quality_cluster_data/rgb_images"
root_folder = "/Users/zetasourpi/cernbox/data/data-for-ml/good_quality_cluster_data/rgb_tensors"

def extract_integer(filename) -> int:
    # expects "object_123.png" or "tensor_123.npz"
    return int(filename.split(".")[0].split("_")[1])

pngs = [f for f in os.listdir(img_folder) if f.endswith(".png")]

npzs = [f for f in os.listdir(root_folder) if f.endswith(".npz")]

print("PNGs:", len(pngs), "NPZs:", len(npzs))

npz_numbers_dict = {extract_integer(f):f for f in npzs}
for png_file in pngs:
    png_number = extract_integer(png_file)
    print(png_number)
    
    img = np.array(Image.open(os.path.join(img_folder, png_file)))

    if png_number not in npz_numbers_dict:
        continue

    npz_file = npz_numbers_dict[png_number]
    data = np.load(os.path.join(root_folder, npz_file))

    print("  PNG:", png_file, "shape:", img.shape, "dtype:", img.dtype)
    print("  NPZ:", npz_file, "keys:", data.files)

    arr = data[data.files[0]]
    print("  NPZ array shape:", arr.shape, "dtype:", arr.dtype)
```

The PNG is definitely not just a raw dump of that tensor — something was done during conversion.
